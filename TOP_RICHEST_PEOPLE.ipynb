##**India Data Set**

```python
x = df.groupby('Country/Territory')
IND = x.get_group('India')
```

```python
IND
```

```python
plt.figure(figsize=(15,23))

plt.subplot(3,1,1)
plt.bar(IND.Industry,IND.NetWorth/10000000000 , width = 0.5)
plt.title("Industries & NetWorth (India)" , size = 20 , color = "#606000")
plt.xlabel('Industry' , size = 15 ,color = "#900050")
plt.ylabel('NetWorth (in Trillion)' , size = 15 ,color = "#900050")

plt.subplot(3,1,2)
plt.bar(IND.Name ,IND.NetWorth/10000000000, width = 0.5)

plt.title('''---------------------------------------------------------------------------------------------------------------------------
          Person & NetWorth (India)''', size = 20 , color = "#606000")
plt.xlabel('Name' , size = 15 ,color = "#900050")
plt.ylabel('NetWorth (in Trillion)' , size = 15 ,color = "#900050")
plt.show()


```

```python

plt.figure(figsize=(28,18))
plt.subplot(1,3,1)
plt.pie(IND.NetWorth, labels = IND.NetWorth,shadow = True , startangle = 90, )
plt.title('Richest Persons in India and their Networth', size = 15 , color = "#900000" )
#plt.legend(IND.Name , title = 'Name of the richest persons in India', loc = 'center' )
plt.subplot(1,3,2)
plt.pie(IND.NetWorth, labels = IND.Industry,shadow = True , startangle = 90 )
plt.title('Richest Persons in India and their Industry', size = 15, color = "#900000" )
plt.legend(IND.Name , title = 'Name of the richest persons in India', loc = 'center' )
plt.subplot(1,3,3)
plt.pie(IND.NetWorth, labels = IND.Source,shadow = True , startangle = 90 )
plt.title('Richest Persons in India and their Source of Income', size = 15 , color = "#900000" )
#plt.legend(IND.Name , title = 'Name of the richest persons in India', loc = 'center' )
plt.show()
```

```python
plt.figure(figsize = (30,5))
plt.subplot(1,3,1)
plt.plot(IND.NetWorth,IND.Age ,color='green', marker='o', linestyle='dashed', linewidth=2, markersize=10 )
plt.grid()
plt.title('Persons age and networth')
plt.xlabel('networth in Trillion')
plt.ylabel('Age')
plt.subplot(1,3,2)
plt.plot(IND.Industry , IND.NetWorth , color='red', marker='v', linestyle='dashed', linewidth=2, markersize=10)
plt.grid()
plt.title('Industry and networth')
plt.ylabel('networth in Trillion')

plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.histplot(IND.Industry , bins = 5 ,)
plt.show()
```

```python
IND.to_csv('India_richest_people.csv')
```
